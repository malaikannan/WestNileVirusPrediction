Converting Categorical variables into dummy variables to be passed as input to Machine Learning Algorithm. I am using Pandas.get_dummies() method, creates a dataframe with different levels in the categorical variable. f_61 has 5 levels a,b,c,d,e hence it creates 5 columns in the f_61_dummies. Dummy variable will have a value of 0 or 1.  Since we only need N-1 dummy variables ( if all N-1 variables value are equal to 0 then the Nth variable value is 1 , hence I am dropping the 1st dummy variable from the dummy variables dataframe. I have to drop the original categorical variable from the training and test dataframe before joining the new dummy dataframe with the original one. 

Transformation applied in Training dataframe has to be applied into testing dataframe too.


```python
#Converting TrainFrame categorical variables to dummy variables
f_61_dummies = pd.get_dummies(credit_card_dataframe['f_61'], prefix='f_61')
f_61_dummies = f_61_dummies.drop('f_61_a',1)
credit_card_dataframe= credit_card_dataframe.drop('f_61',1)
credit_card_dataframe = credit_card_dataframe.join(f_61_dummies.ix[:,:])

f_121_dummies = pd.get_dummies(credit_card_dataframe['f_121'], prefix='f_121')
f_121_dummies = f_121_dummies.drop('f_121_A',1)
credit_card_dataframe= credit_card_dataframe.drop('f_121',1)
credit_card_dataframe = credit_card_dataframe.join(f_121_dummies.ix[:,:])

f_215_dummies = pd.get_dummies(credit_card_dataframe['f_215'], prefix='f_215')
f_215_dummies = f_215_dummies.drop('f_215_blue',1)
credit_card_dataframe= credit_card_dataframe.drop('f_215',1)
credit_card_dataframe = credit_card_dataframe.join(f_215_dummies.ix[:,:])

f_237_dummies = pd.get_dummies(credit_card_dataframe['f_237'], prefix='f_237')
f_237_dummies = f_237_dummies.drop('f_237_Canada',1)
credit_card_dataframe= credit_card_dataframe.drop('f_237',1)
credit_card_dataframe = credit_card_dataframe.join(f_237_dummies.ix[:,:])
```

```python
#Converting TestFrame categorical variables to dummy variables
f_61_test_dummies = pd.get_dummies(credit_card_testframe['f_61'], prefix='f_61')
f_61_test_dummies = f_61_test_dummies.drop('f_61_a',1)
credit_card_testframe= credit_card_testframe.drop('f_61',1)
credit_card_testframe = credit_card_testframe.join(f_61_test_dummies.ix[:,:])

f_121_test_dummies = pd.get_dummies(credit_card_testframe['f_121'], prefix='f_121')
f_121_test_dummies = f_121_test_dummies.drop('f_121_A',1)
credit_card_testframe= credit_card_testframe.drop('f_121',1)
credit_card_testframe = credit_card_testframe.join(f_121_test_dummies.ix[:,:])

f_215_test_dummies = pd.get_dummies(credit_card_testframe['f_215'], prefix='f_215')
f_215_test_dummies = f_215_test_dummies.drop('f_215_blue',1)
credit_card_testframe= credit_card_testframe.drop('f_215',1)
credit_card_testframe = credit_card_testframe.join(f_215_test_dummies.ix[:,:])

f_237_test_dummies = pd.get_dummies(credit_card_testframe['f_237'], prefix='f_237')
f_237_test_dummies = f_237_test_dummies.drop('f_237_Canada',1)
credit_card_testframe= credit_card_testframe.drop('f_237',1)
credit_card_testframe = credit_card_testframe.join(f_237_test_dummies.ix[:,:])
```

Missing values in the other dataframe columns have to be handled. I am using sklearn Imputer with the strategy of fill it with Mean to fill the missing values. Imputer will give o/p in numpyarray format.

Applying the transformation in both Training and Test Dataframe.

```python
from sklearn.preprocessing import Imputer

 
imp = Imputer(missing_values='NaN', strategy='mean', axis=0)
imp.fit(credit_card_dataframe)
credit_card_dataframe_imp = imp.transform(credit_card_dataframe)

imp.fit(credit_card_testframe)
credit_card_testframe_imp = imp.transform(credit_card_testframe)

#numpy.savetxt("C:/Ipython/Capitalone/Data/codetest_2083.tar/submission.csv",credit_card_testframe_imp,delimiter=",")
```
